#### sMAPE и Final sMAPE

Напишем функции для расчета клюевых показателей в процессе очистки руды

```python
def smape (target, predict):
    return ((1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100)
```

```python
def final_smape (smape_rougher, smape_final):
    return 0.25*smape_rougher+0.75*smape_final
```

```python
smape_scorer = make_scorer(smape)
```

#### Обучение модели LinearRegression

#### Обучение модели DecisionTreeRegressor

```python
%%time 
best_model_rougher_dtr = None
best_model_final_dtr = None
best_final_smape_dtr = 1000000
depth_dtr = 0
list_depth_dtr = []

for i in range (1,10): #изначаьно взял на пробуу глубину 100, но сокралит, до 10, т.к. нашел лучший рез-т и лишнее время на обработу тратить ни к чему
    model_rougher_dtr = DecisionTreeRegressor(random_state=45, max_depth = i) # инициализируем модель DecisionTreeRegressor для этапа обогощения
    model_rougher_dtr.fit(features_rougher, target_rougher) # обучаем модель на тренировочной выборке для этапа обогощения
    smape_rougher_dtr = cross_val_score(model_rougher_dtr, features_rougher, target_rougher, scoring=smape_scorer, cv=7)
    mean_smape_rougher_dtr = smape_rougher_dtr.mean()

    
    model_final_dtr = DecisionTreeRegressor(random_state=45, max_depth = i) # инициализируем модель DecisionTreeRegressor для финала
    model_final_dtr.fit(features_final, target_final) # обучаем модель на тренировочной выборке для финала
    smape_final_dtr = cross_val_score(model_final_dtr, features_final, target_final, scoring = smape_scorer, cv=7)
    mean_smape_final_dtr = smape_final_dtr.mean()#берем среднее от получившегося списка результатов    
    
    final_smape_dtr = final_smape(mean_smape_rougher_dtr, mean_smape_final_dtr)
    list_depth_dtr.append(final_smape_dtr)
     
    if final_smape_dtr < best_final_smape_dtr:
        best_final_smape_dtr = final_smape_dtr
        best_model_rougher_dtr = model_rougher_dtr
        best_model_final_dtr = model_final_dtr
        depth_dtr = i
print(f'Лучший final_smape:', best_final_smape_dtr, "при глубине дерева:", depth_dtr)        
```

#### Обучение модели RandomForestRegressor

```python
%%time 
best_model_rougher_rfr = None
best_model_final_rfr = None
best_final_smape_rfr = 1000000
depth_rfr = 0
est_rfr = 0
for est in range (1,3):# изначально задал оба цикла range(3,12,3). Здесь пердставлен демонтарционный итог 
    for i in range (1,3): #изначаьно взял на пробуу глубину range(3,12,3), но сокралит, до 3, т.к. лучший рез-т оказался 1, как и в DTR лишнее время на обработу тратить ни к чему
        model_rougher_rfr = DecisionTreeRegressor(random_state=45, max_depth = i) # инициализируем модель RF для этапа обогощения
        model_rougher_rfr.fit(features_rougher, target_rougher) # обучаем модель на тренировочной выборке для этапа обогощения
        smape_rougher_rfr = cross_val_score(model_rougher_rfr, features_rougher, target_rougher, scoring=smape_scorer, cv=7)
        mean_smape_rougher_rfr = smape_rougher_rfr.mean()


        model_final_rfr = DecisionTreeRegressor(random_state=45, max_depth = i) # инициализируем модель DecisionTreeRegressor для финала
        model_final_rfr.fit(features_final, target_final) # обучаем модель на тренировочной выборке для финала
        smape_final_rfr = cross_val_score(model_final_rfr, features_final, target_final, scoring = smape_scorer, cv=7)
        mean_smape_final_rfr = smape_final_rfr.mean()#берем среднее от получившегося списка результатов    
    
        final_smape_rfr = final_smape(mean_smape_rougher_rfr, mean_smape_final_rfr)
    
        if final_smape_rfr < best_final_smape_rfr:
            best_final_smape_rfr = final_smape_rfr
            best_model_rougher_rfr = model_rougher_rfr
            best_model_final_rfr = model_final_rfr
            depth_rfr = i
            est_rfr = est
print(f'Лучший final_smape:', best_final_smape_rfr, "при глубине дерева:", depth_rfr, 'и кол-ве деревьев:', est_rfr)        
```
